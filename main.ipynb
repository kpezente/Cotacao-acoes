```python
#Importações dos módulos utilizados
import requests
import pandas as pd
import pprint
import os
from io import StringIO
```

```python
#Buscando a chave da API
pasta = r'C:\PYTHON Hashtag\33. Integracao Python com APIs e JSON\5. API Ações'
caminho = os.path.join(pasta, "chave.py")
print(caminho)
with open(caminho, "r") as arquivo:
  api_key = arquivo.read()
```

```python
#Function para adicionar uma ação.
def adicionar_acao(lista):
  acao = input('Informe a ação a ser adicionado:')
  acao = acao.upper()
  lista.append(acao)
  print(f'{acao}: adicionada com sucesso.')
  return acao


#Function para remover uma ação.
def remover_acao(lista):
  #Verificando se a lista possui alguma ação.
  if not lista:
    print('A lista de ações está vazia')
  else:
    acao = input('Informe a ação a ser removido:')
    acao = acao.upper()
    if acao in lista:
      lista.remove(acao)
      print(f'{acao}: removido com sucesso')
    else:
      print(f'A ação: {acao} não está na lista. ')

#Function para ver a tabela com a lista de ações.
def ver_lista(lista, compilada):
  if lista:
    for acao in lista:
      url = f'https://www.alphavantage.co/query?function=GLOBAL_QUOTE&symbol={acao}.SAO&apikey={api_key}&datatype=csv'
      r = requests.get(url)
      tabela = pd.read_csv(StringIO(r.text))
      lista_tabelas = [compilada, tabela]
      compilada = pd.concat(lista_tabelas)
    display(compilada)  
  else:
    print('A lista de ações está vazia.')
  return compilada

#Function para procurar o código de uma ação pelo nome da empresa.
def procurar_acao():
  empresa = input('Informe a empresa que deseja procura:')
  print(empresa)
  url = f'https://www.alphavantage.co/query?function=SYMBOL_SEARCH&keywords={empresa}&apikey={api_key}&datatype=csv'
  r = requests.get(url)
  tabela = pd.read_csv(StringIO(r.text))
  display(tabela)

#Function para ver o histórico diário de uma ação atualizado a cada 10 min.
def historico_diario_10():
  acao = input('Informe a ação que deseja consultar o histórico:')
  url = f'https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY&symbol={acao}&interval=10min&apikey={api_key}&datatype=csv'
  r = requests.get(url)
  tabela = pd.read_csv(StringIO(r.text))
  display(tabela)
  

#Function para ver o histórico semanal de uma ação.
def historico_semanal():
  acao = input('Informe a ação que deseja consultar o histórico:')
  url = f'https://www.alphavantage.co/query?function=TIME_SERIES_WEEKLY_ADJUSTED&symbol={acao}&apikey={api_key}&datatype=csv'
  r = requests.get(url)
  tabela = pd.read_csv(StringIO(r.text))
  display(tabela)

#Function do menu.
def main():
  return int(input('Escolha a opção desejada:' '\n'
                 '1 = Adicionar uma ação' '\n'
                 '2 = Remover uma ação' '\n'
                 '3 = Procurar uma ação' '\n'
                 '4 = Histórico do dia atualizado a cada 10min de uma ação\n'
                 '5 = Ver histórico dos últimos 30 dias de uma ação\n'
                 '6 = Ver histórico semanal de uma ação\n'
                 '7 = Sair\n'
                 ))
```

```python
opcao = main()
lista_acoes = []
compilada = pd.DataFrame()

#Rodando o código em looping até solicitar a saída.
while opcao != 7:
  if opcao == 1:
    adicionar_acao(lista_acoes)
  elif opcao == 2:
    remover_acao(lista_acoes)
  elif opcao == 3:
    ver_lista(lista_acoes, compilada)
  elif opcao == 4:
    procurar_acao()
  elif opcao == 5:
    historico_diario_10()
  elif opcao == 6:
    historico_semanal()
  else:
    print('Opcão não encontrada. Digite novamente.')

  opcao = main()


```
